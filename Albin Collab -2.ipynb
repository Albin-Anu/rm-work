```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def display(self):
        print(f"Name: {self.name}, Grade: {self.grade}")

# Create 2 student objects
student1 = Student("Albin", "A")
student2 = Student("Anu", "B")

# Display details of each student
student1.display()
student2.display()
```

```python
class Consumer:
    def __init__(self, name, city):
        self.name = name
        self.city = city

    def display(self):
        print(f"Name: {self.name}, City: {self.city}")


# Create 5 consumer objects
consumer1 = Consumer("Ramu", "Mumbai")
consumer2 = Consumer("Devajith", "Delhi")
consumer3 = Consumer("John", "Chennai")
consumer4 = Consumer("kate", "Bangalore")
consumer5 = Consumer("Kiran", "Kolkata")

# Display details of each consumer
consumer1.display()
consumer2.display()
consumer3.display()
consumer4.display()
consumer5.display()

```
